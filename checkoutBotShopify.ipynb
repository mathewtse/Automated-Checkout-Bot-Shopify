```python
# first import the libraries and open the csv file

from selenium import webdriver as wd
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.remote.webelement import WebElement

import chromedriver_binary
import random
import time

wd = wd.Chrome(executable_path=r'C:\bin\chromedriver.exe')
wd.implicitly_wait(10)

wd.get("https://www.inbtwncm.com/4131684421/checkouts/8316bca9f52a041c46d2d50bc5b13caa?_ga=2.5968755.1791437129.1653354859-1019354792.1653354859")
wd.maximize_window()

def randomSleep():
    random_wait_time = random.randrange(5.0,10)
    print(random_wait_time)
    time.sleep(random_wait_time)
    
def randomSleepLong():
    random_wait_time = random.randrange(12,18)
    print(random_wait_time)
    time.sleep(random_wait_time)

def slow_type(element, text):
    for character in text:
        delay = random.randrange(0,100)/100
        element.send_keys(character)
        time.sleep(delay)

file = open("bot-fulfillment-info.csv")
   

#CREDIT CARD INFORMATION

cardNumberValue = "1234123412341234"
expiryDateValue = "1234"
securityNumValue = "440"
```

```python
# then loop through the entire csv file


for line in file:
    lineList = line.strip().split(",")
    
    emailValue = lineList[0] 
    firstNameValue = lineList[1]
    lastNameValue = lineList[2]
    addressLineOneValue = lineList[3]
    addressLineTwoValue = lineList[4]
    cityValue = lineList[5]
    countryValue = lineList[6]
    stateValue = lineList[7]
    zipCodeValue = lineList[8]
    phoneNumberValue = lineList[9]
  
    ## first page of the checkout
    

    randomSleep()
    
    email = wd.find_element_by_xpath('//*[@id="checkout_email"]')
    type(email)
    slow_type(email,emailValue)
    randomSleep()

    firstName = wd.find_element_by_xpath('//*[@id="checkout_shipping_address_first_name"]')
    type(firstName)
    slow_type(firstName,firstNameValue)
    randomSleep()

    lastName = wd.find_element_by_xpath('//*[@id="checkout_shipping_address_last_name"]')
    type(lastName)
    slow_type(lastName,lastNameValue)
    randomSleep()

    address = wd.find_element_by_xpath('//*[@id="checkout_shipping_address_address1"]')
    type(address)
    slow_type(address,addressLineOneValue)
    randomSleep()

    addressLineTwo = wd.find_element_by_xpath('//*[@id="checkout_shipping_address_address2"]') 
    type(addressLineTwo)
    slow_type(addressLineTwo,addressLineTwoValue)
    randomSleep()

    city = wd.find_element_by_xpath('//*[@id="checkout_shipping_address_city"]')
    type(city)
    slow_type(city,cityValue)
    randomSleep()

    country = wd.find_element_by_xpath('//*[@id="checkout_shipping_address_country"]')
    Select(country).select_by_value('United States')
    randomSleep()

    state = wd.find_element_by_xpath('//*[@id="checkout_shipping_address_province"]')
    Select(state).select_by_value(stateValue)
    randomSleep()

    zipCode = wd.find_element_by_xpath('//*[@id="checkout_shipping_address_zip"]')
    type(zipCode)
    slow_type(zipCode, zipCodeValue)
    randomSleep()
    
    phoneNumber = wd.find_element_by_xpath('//*[@id="checkout_shipping_address_phone"]')
    type(phoneNumber)
    slow_type(phoneNumber, phoneNumberValue)
    randomSleep()

    discountCodeInput = wd.find_element_by_xpath('//*[@id="checkout_reduction_code"]')
    type(discountCodeInput)
    discountCodeInput.send_keys("SAVE15")
    randomSleep()

    discountCodeBtn = wd.find_element_by_xpath('//*[@id="checkout_submit"]')
    discountCodeBtn.click()
    randomSleep()
        
    continueToShippingBtn = wd.find_element_by_xpath('//*[@id="continue_button"]')
    continueToShippingBtn.click() 
    randomSleepLong()
    
    ## now on second page of checkout
    
    
    continueToPaymentBtn = wd.find_element_by_xpath('//*[@id="continue_button"]')
    continueToPaymentBtn.click()
    randomSleepLong()
    
    ## now on third page of checkout
    
        
    wd.switch_to.default_content()

    iframe = wd.find_element_by_class_name('card-fields-iframe')
    wd.switch_to.frame(iframe)
    cardNumber = wd.find_element_by_name('number')
    time.sleep(1)
    slow_type(cardNumber,cardNumberValue)

    
    ##inputting name on card
    
    nameOnCard = ActionChains(wd) 
    for _ in range(2):
        nameOnCard = nameOnCard.send_keys(Keys.TAB)

    for char in firstNameValue:
        nameOnCard = nameOnCard.pause(random.random())
        nameOnCard = nameOnCard.send_keys(char)

    nameOnCard = nameOnCard.send_keys(" ")

    for char in lastNameValue:
        nameOnCard = nameOnCard.pause(random.random())
        nameOnCard = nameOnCard.send_keys(char)

    nameOnCard.pause(random.random())

    nameOnCard.perform()

    randomSleep()


    ## inputting expiry date

    expirationDate = ActionChains(wd)
    expirationDate = expirationDate.send_keys(Keys.TAB)

    for char in expiryDateValue:
        expirationDate = expirationDate.pause(random.random())
        expirationDate = expirationDate.send_keys(char)

    expirationDate.perform()

    randomSleep()

    #inputting the security code

    securityCode = ActionChains(wd)
    securityCode = securityCode.send_keys(Keys.TAB)

    for char in securityNumValue:
        securityCode.pause(random.random())
        securityCode = securityCode.send_keys(char)

    securityCode.perform()

    randomSleep()

    wd.switch_to.default_content()

    payNowBtn = wd.find_element_by_xpath('//*[@id="continue_button"]')
    payNowBtn.click()

    waitTimeUntilNextFulfillment = random.randrange(20,60) # next order is executed sometime between 20-60 minutes
    print("the wait time in minutes until the next fulfillment is " + str(waitTimeUntilNextFulfillment))
    time.sleep(waitTimeUntilNextFulfillment * 60)
    
    wd.get("https://www.inbtwncm.com/4131684421/checkouts/8316bca9f52a041c46d2d50bc5b13caa?_ga=2.5968755.1791437129.1653354859-1019354792.1653354859")
```
